```python
import torch
torch.set_default_device('cpu')
torch.set_default_dtype(torch.float64)
import sys
sys.path.append("../")
from simulations.simulator import wiener
import matplotlib.pyplot as plt
from model.gp_train_batch import train_gp
from model.models import DeepGP_Wiener
from dynonetloc.dynonet_train import train_dynonet
import numpy as np
import matplotlib.patches as mpatches
```

```python
deterministic = False
torch.manual_seed(2244)    #142

N = 2200  # total number of data points
N_train = 1500  #
deltat = 1 / 100.0  # sampling rate

number_in_b = 2  # number of inputs - they are shifted cosine functions
num_states = 5
number_in_l = 1  # number of columns of the matrix L which defines the covariance of the noise as LL^T

# add measreument noise
noisestd = 0.02

# static nonlinearity
def nonlinear(x):
    return x**2 #(x>0)*x #(x>0)*x  #x**2  # torch.sigmoid(x)

t, u, y, parameters = wiener(
        deltat,
        N,
        N_train,
        num_states=num_states,
        number_in_b=number_in_b,
        number_in_l=1,
        jitter=1e-8,
        nonlinear=nonlinear,
        deterministic=deterministic,
    )  # torch.exp)
    
# normalise data for fitting
y = (y - y.mean()) / y.std()
u = (u - u.mean(axis=1)) / u.std(axis=1)

#add noise
y = y + noisestd * torch.randn(*y.shape)


fig,ax=plt.subplots(figsize=(12,4))
ax.scatter(t.flatten(),y.flatten(),color='C1',s=0.8,zorder=1e6,label="data")
plt.plot(t.flatten(),y.flatten(),color='C1',alpha=0.3)
plt.legend()
plt.ylabel("y",fontsize=14)
plt.xlabel("time",fontsize=14)
#ax.axvline(time_tr[-1],color='black')
r_patch = mpatches.Patch(color='C1', label='data',linewidth=0.1)
plt.legend(handles=[r_patch])
#'plt.tight_layout()
#plt.savefig("data_x2.pdf")

#deterministic = False  # if this is set to True then the linear system is deterministic
```

```python

time_tr = t[0:N_train, :]
time_te = t
train_x = torch.cat(
    [time_tr, u[0, 0:N_train, :]], dim=1
)  # the input is the stacking of time and u [t,u]
train_y = y[0:N_train]
train_u = u[:, :N_train, :] 

# we test it on all the data
test_x = torch.cat([time_te, u[0, ...]], dim=1)
test_y = y


shiftp0 = 500
shiftp = np.minimum(shiftp0,N_train)
test_x = test_x[N_train-shiftp:,...]
test_y = test_y[N_train-shiftp:,...]
test_u = u[:,N_train-shiftp:, ...]


```

```python
plt.plot(test_x[:,0].flatten(),test_x[:,1].flatten(),label="dynoNet")
plt.plot(test_x[:,0].flatten(),test_x[:,2].flatten(),label="dynoNet")
```
