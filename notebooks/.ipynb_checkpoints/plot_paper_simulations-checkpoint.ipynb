```python
import torch
torch.set_default_device('cpu')
torch.set_default_dtype(torch.float64)
import sys
sys.path.append("../")
from simulations.simulator import wiener
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.patches as mpatches
import pickle
import pandas as pd
```

```python
path = "../results/WinerHammersteinReal_stored/wiener_hamemrstein"
with open(path+"data", 'rb') as f:
    data = pickle.load(f)
with open(path+"gp_means", 'rb') as f:
    predictive_means = pickle.load(f)
with open(path+"gp_vars", 'rb') as f:
    predictive_variances = pickle.load(f)
dNN = pd.read_csv(path+"dyno.csv",header=None)
dNN.shape

t_tr= data[0][:,0].numpy()
u_tr= data[0][:,1].numpy()
t_te= data[2][:,0].numpy()
u_te= data[2][:,1].numpy()
y_tr= data[1].numpy()
y_te= data[3].numpy()

test_y_hat_NN = dNN.values.flatten()

```

```python



fig,ax=plt.subplots(figsize=(12,4))
ax.scatter(t_tr.flatten(),y_tr.flatten(),color='C1',s=0.8,zorder=1e6,label="data")
plt.plot(t_tr.flatten(),y_tr.flatten(),color='C1',alpha=0.3)
plt.legend()
plt.ylabel("y",fontsize=14)
plt.xlabel("time",fontsize=14)
#ax.axvline(time_tr[-1],color='black')
r_patch = mpatches.Patch(color='C1', label='data',linewidth=0.1)
plt.legend(handles=[r_patch])
```

```python
fig,ax=plt.subplots(figsize=(12,4))
time_tep = t_te
ax.scatter(time_tep.flatten(),y_te.flatten(),color='C1',s=0.8,zorder=1e6,label="data")
ax.plot(time_tep.flatten(),test_y_hat_NN,label="dynoNet",color='C2')
ax.axvline(t_tr[-1],color='black')
plt.ylabel("y",fontsize=14)
plt.xlabel("time",fontsize=14)
r_patch = mpatches.Patch(color='C1', label='data',linewidth=0.1)
b_patch = mpatches.Patch(color='C2', label='dynoNet')
plt.legend(handles=[r_patch,b_patch])
#ax.set_xlim([10.1,22.5])
#ax.set_ylim([-1.5,6.0])
#plt.savefig("dynoNet_x2.pdf")
```

```python
from functions.plotting import plot_gp_dist
fig,ax=plt.subplots(figsize=(36,9))
ax.scatter(time_tep.flatten(),y_te.flatten(),color='C1',s=0.8,zorder=1e6,label="data")
#meanp = predictive_means.mean(axis=0).flatten()
samples=[]
nsamples=200
for i in range(nsamples):
    j = np.random.randint(predictive_means.shape[0])
    s = predictive_means[j,0,:]+np.sqrt(predictive_variances[j,0,:])*np.random.randn(predictive_variances.shape[2])
    samples.append(s)
samples = np.vstack(samples)
plot_gp_dist(ax,samples,time_tep.flatten(),samples_alpha=0.3,palette='Blues',label=None)
ax.plot(time_tep.flatten(),np.median(predictive_means,axis=0).flatten(),'darkblue')
#plot_gp_dist(ax,predictive_means[:,0,:],time_tep.flatten(),samples_alpha=0.01,palette='Blues',label=None)

ax.axvline(t_tr[-1],color='black')
ax.set_ylabel("y",fontsize=14)
ax.set_xlabel("time",fontsize=14)

r_patch = mpatches.Patch(color='C1', label='data',linewidth=0.1)
b_patch = mpatches.Patch(color='darkblue', label='dynoGP')
plt.legend(handles=[r_patch,b_patch])
#plt.tight_layout()
plt.savefig("dynoGP_x22.pdf")
```
